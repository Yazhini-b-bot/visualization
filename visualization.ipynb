```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Product":["Laptop","Phone","Tab","Monitor"],
    "Month":["Jan","Feb","Mar","Apr"],
    "Quantity":[10,20,15,25],
    "Sales":[45000,18000,25000,35000]
}
df=pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(6,5))
plt.bar(df["Product"],df["Sales"],color=["blue","green","yellow","purple"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.plot(
    df["Month"],
    df["Sales"],
    marker="o",
    linestyle="-",
    linewidth=2
)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Analysis")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df["Sales"],labels=df["Product"],autopct="%1.1f%%")
plt.title("Sales Analysis")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.bar(df["Product"],df["Sales"],color=["blue","green","yellow","purple"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales Analysis")

plt.subplot(1,3,2)
plt.plot(df["Product"],df["Sales"],marker="o",color="purple")
plt.xlabel("Product")
plt.ylabel("Profit")
plt.title("Sales Trend")

plt.subplot(1,3,3)
plt.pie(df["Sales"],labels=df["Product"],autopct="%1.1f%%")
plt.title("Proportion")
plt.show()

plt.tight_layout()
plt.show()
```
